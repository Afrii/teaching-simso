# Linear and Exponential Growth

Linear growth:

$$ \frac{dp}{dt} = \lambda $$

$$ p(t) = \lambda t + \hbox{const} $$

Exponential growth:

$$ \frac{dp}{dt} = \lambda p $$

$$ p(t) = e^{\lambda t} + \hbox{const} $$

Combined growth:

$$ \frac{dx}{dt} = ax + b $$

$$ x(t) = \left( x(0) + \frac{b}{a} \right) e^{at} - \frac{b}{a} $$

Taylor theorem:

$$ \frac{dx}{dt} = f(x) $$

$$ f(x) \approx f(x_0) + f'(x_0) (x-x_0) + O(\delta^2) $$

$$ \frac{dx}{dt} \approx f(x_0) + f'(x_0)(x-x_0) $$

To first order, growth is exponential over short time spans.

# Logistic Function

Population dependent growth rate $h(p)$:

$$ \frac{dp}{dt} = h(p) p $$

Growth rate is difference between birth rate and death rate.

$$ h(p) = k - rp $$

Here, $k$ is the birth rate, and $r$ is a death rate resulting from overpopulation.

$$ \frac{dp}{dt} = kp - rp^2 $$

This gives rise to the _logistic regression_.

Stable Points:

$$ \frac{dx}{dt} = f(x) $$

Where $f(x)=0$, the solution is stationary.

Where $f(x)=0$ and $f'(x)&lt;0$, the stationary solution is stable, because if $x$ is too small, it causes $x$ to increase, and if $x$ is too large, it causes $x$ to decrease.

# Systems of First Order Equations

Consider two types of snails, left and right handed. 

- Assume their numbers are $L$ and $R$.
- We assume a logistic-style equation (competition for limited resources).
- Left and right snails interact with some probability and change into the opposite kind.

$$ \frac{dR}{dt} = R - (R^2 + a RL) $$
$$ \frac{dL}{dt} = L - (L^2 + a LR) $$

Observation on Snails:

- real snail populations are almost entirely left-handed or right-handed
- yet, there is no known advantage to either handedness

Why does one handedness or the other win?

Nullclines:

Look at equations for $\frac{dR}{dt}=0$ and $\frac{dL}{dt}=0$.


$ R - (R^2 + aRL) = 0 $ or $ R (1 - R - aL) = 0 $ 

$$ R = 0 \vee R = 1-aL $$
$$ L = 0 \vee L = 1-aR $$


```python
figsize(8,8)
L,R = meshgrid(linspace(0,1,20),linspace(0,1,20))
a=2; dL = L-(L**2+a*R*L); dR = R-(R**2+a*R*L); n = hypot(dL,dR)
quiver(L,R,dL/n,dR/n); plot([0,0.5],[1,0]); plot([0,1],[0.5,0])
```

# Harmonic Oscillator and Linear Systems

The linear system in two variables $x$, $y$ is the _harmonic oscillator_:

$$ \frac{dx}{dt} = -y $$
$$ \frac{dy}{dt} = x $$

This is equivalent (trivially) to the second order differential equation:

$$ \frac{d^2x}{dt^2} = -x $$

We can transform higher order differential equations into first order equations by introducing such additional coordinates.

Physical model:

Consider a spring and Hooke's law:

$$ F = k x $$

Newton's second law:

$ F = m a $ and $ a = \ddot x $

$$ F = k x = - m a = - m \ddot x $$

Differential equation:

$$ \ddot x = - \frac{k}{m} x $$

```python
def odeanalyze(f,phase=(-1,-1,1,1),time=(0,20),start=(1.0,0.0),ma=False):
    figsize(12,6)
    x,y = meshgrid(linspace(phase[0],phase[2],100),linspace(phase[1],phase[3],100))
    dx,dy = f(x,y)
    subplot(121)
    if ma: mathaxes()
    streamplot(x,y,dx,dy)
    contour(x,y,dx,levels=[0],linewidths=3,colors='yellow')
    contour(x,y,dy,levels=[0],linewidths=3,colors='orange')
    def g(x,t): return f(x[0],x[1])
    t = linspace(time[0],time[1],1000)
    x = odeint(g,array(start),t)
    subplot(122)
    if ma: mathaxes()
    plot(t,x[:,0]); plot(t,x[:,1])
```

```python
def f(x,y): return (-y,x)
odeanalyze(f)
```

```python
def f(x,y): return (x,0.5*y)
odeanalyze(f)
```

Stability criteria:

- look at the eigenvalues of $M$
- if they are both real, we have exponential growth
- if they are both imaginary, we have periodic solutions

# Lotka-Volterra Equation

Foxes vs hares:

$$ \frac{dH}{dt} = (a - bH - cF) H $$

$$ \frac{dF}{dt} = (-d + eH) F $$

Simple constants for Lotka-Volterra:

$$ \frac{dH}{dt} = (2 - H - F) H $$

$$ \frac{dF}{dt} = (-1 + H) F $$


```python
def f(H,F): return ((2-H-F)*H,(-1+H)*F)
odeanalyze(f,phase=(0,0,2,2),start=(0.1,0.1))
```

```python
def f(H,F): return (H*(1-F),-0.3*F*(1-H))
odeanalyze(f,phase=(0,0,2,2),start=(0.1,0.1),time=(0,70))
```

# FitzHugh Nagumo Model

The FitzHugh-Nagumo modelis a simplified neuron model consisting of:

- a voltage-like variable $v$
- an inhibitory variable $w$
- a time-dependent current $I_e(t)$

$$ \dot{v} = v - v^3/3 - w + I_e(t) $$
$$ \tau \dot{w} = v - a - bw $$

```python
def fhm(v,w,a=0.0,b=0.0,tau=1.0,t=100,dt=1e-3,ie=lambda t:0.0):
    dv = (v-v**3/3.0-w+ie(t))*dt
    dw = ((v-a-b*w)/tau)*dt
    return dv,dw
odeanalyze(fhm,phase=(-4,-4,4,4),start=(0,-3),time=(0,30000),ma=1)
```

# van der Pol Oscillator and Limit Cycles

$$\dot x = y$$
$$\dot y = \mu(1-x^2) y-x.$$

```python
mu = 0.5
def vdp(x,y):
    dx = y
    dy = mu*(1-x**2)*y-x
    return dx,dy
odeanalyze(vdp,phase=(-6,-6,6,6),start=(1,1),time=(0,40),ma=1)
```

Limit cycle:

A limit cycle is an _isolated periodic orbit_.

# Chaos

```python
mu = 0.5
def g(v,t):
    x,y = list(v)
    return y*(1+0.3*sin(1.73*t)),mu*(1-x**2)*y-x
x = odeint(g,(0.1,0.1),t)
plot(x[5000:,0],x[5000:,1],color='blue')
def f(v,t):
    x,y = list(v)
    return y,mu*(1-x**2)*y-x
t = linspace(0.0,1000.0,10000)
x = odeint(f,(0.1,0.1),t)
plot(x[5000:,0],x[5000:,1],color='red')
```

```python
def f(v,t,sigma=10,beta=8.0/3,rho=28.0):
    x,y,z = list(v)
    return sigma *( y - x ),x *( rho - z ) -y,x *y - beta * z
t = linspace(0.0,50.0,10000)
x = odeint(f,(0.1,0.1,0.1),t)
plot(t,x)
```

# Summary - Differential Equations

Classification of differential equations:

- ordinary vs partial
- linear vs non-linear

Phase space:

- plot flow field of derivatives
- look for nullclines
- look for asymptotic behavior
- look for limit cycles

Stability:

- in 1D, $f(x)=0$ and $f'(x)&lt;0$
- in 2D, $\dot x = 0$ and $\dot y = 0$ are stationary points
- stability given by the surrounding nullclines, eigenvalues of the Jacobian

Modeling:

- differential equations are usually continuous approximations to discrete systems
