# Linear and Exponential Growth

Linear growth:

$$ \frac{dp}{dt} = \lambda $$

$$ p(t) = \lambda t + \hbox{const} $$

Exponential growth:

$$ \frac{dp}{dt} = \lambda p $$

$$ p(t) = e^{\lambda t} + \hbox{const} $$

Combined growth:

$$ \frac{dx}{dt} = ax + b $$

$$ x(t) = \left( x(0) + \frac{b}{a} \right) e^{at} - \frac{b}{a} $$

Taylor theorem:

$$ \frac{dx}{dt} = f(x) $$

$$ f(x) \approx f(x_0) + f'(x_0) (x-x_0) + O(\delta^2) $$

$$ \frac{dx}{dt} \approx f(x_0) + f'(x_0)(x-x_0) $$

To first order, growth is exponential over short time spans.

# Logistic Function

Population dependent growth rate $h(p)$:

$$ \frac{dp}{dt} = h(p) p $$

Growth rate is difference between birth rate and death rate.

$$ h(p) = k - rp $$

Here, $k$ is the birth rate, and $r$ is a death rate resulting from overpopulation.

$$ \frac{dp}{dt} = kp - rp^2 = p (k - rp)$$

This gives rise to the _logistic function.

```python
def f(x,t): return 0.001*(500-x)*x
t = linspace(0.0,50.0,10000)
x = odeint(f,1e-3,t)
plot(t,x)
```

```python
for start in linspace(0.0,1000.0,10):
    t = linspace(0.0,50.0,10000)
    x = odeint(f,start,t)
    plot(t,x)
```

```python
for start in exp(linspace(log(1e-6),log(1e3),20)):
    t = linspace(0.0,50.0,10000)
    x = odeint(f,start,t)
    plot(t,x)
```

We can interpret this as two exponential growth regions:

- $\dot x = \alpha x$ near $x=0$ (exponential growth)
- $\dot x = c-\beta x$ near $x=500$ (stable point with exponential decay)

```python
mathaxes(); x = linspace(-100,600,1000)
plot(x,f(x,0)); plot(x[x<100],0.001*500*x[x<100]); plot(x[x>400],250+0.001*-500*x[x>400])
```

Stable Points:

$$ \frac{dx}{dt} = f(x) $$

Where $f(x)=0$, the solution is stationary.

Where $f(x)=0$ and $f'(x)&lt;0$, the stationary solution is stable, because if $x$ is too small, it causes $x$ to increase, and if $x$ is too large, it causes $x$ to decrease.

# Systems of First Order Equations

Consider two types of snails, left and right handed. 

- Assume their numbers are $L$ and $R$.
- We assume a logistic-style equation (competition for limited resources).
- Left and right snails interact with some probability and change into the opposite kind.

$$ \frac{dR}{dt} = R - (R^2 + a RL) $$
$$ \frac{dL}{dt} = L - (L^2 + a LR) $$

Observation on Snails:

- real snail populations are almost entirely left-handed or right-handed
- yet, there is no known advantage to either handedness

Why does one handedness or the other win?

Nullclines:

Look at equations for $\frac{dR}{dt}=0$ and $\frac{dL}{dt}=0$.


$ R - (R^2 + aRL) = 0 $ or $ R (1 - R - aL) = 0 $ 

$$ R = 0 \vee R = 1-aL $$
$$ L = 0 \vee L = 1-aR $$


```python
figsize(8,8)
L,R = meshgrid(linspace(0,1,20),linspace(0,1,20))
a=2; dL = L-(L**2+a*R*L); dR = R-(R**2+a*R*L); n = hypot(dL,dR)
quiver(L,R,dL/n,dR/n); plot([0,0.5],[1,0]); plot([0,1],[0.5,0])
```

# Harmonic Oscillator and Linear Systems

The linear system in two variables $x$, $y$ is the _harmonic oscillator_:

$$ \frac{dx}{dt} = -y $$
$$ \frac{dy}{dt} = x $$

This is equivalent (trivially) to the second order differential equation:

$$ \frac{d^2x}{dt^2} = -x $$

We can transform higher order differential equations into first order equations by introducing such additional coordinates.

Physical model:

Consider a spring and Hooke's law:

$$ F = k x $$

Newton's second law:

$ F = m a $ and $ a = \ddot x $

$$ F = k x = - m a = - m \ddot x $$

Differential equation:

$$ \ddot x = - \frac{k}{m} x $$

```python

def odeanalyze(f,phase=(-1,-1,1,1),time=(0,20),start=(1.0,0.0),ma=False,limit=None,ltime=None,lN=10000,lstart=None):
    figsize(12,6)
    x,y = meshgrid(linspace(phase[0],phase[2],100),linspace(phase[1],phase[3],100))
    dx,dy = f(x,y)
    subplot(121)
    if ma: mathaxes()
    streamplot(x,y,dx,dy)
    contour(x,y,dx,levels=[0],linewidths=3,colors='yellow')
    contour(x,y,dy,levels=[0],linewidths=3,colors='orange')
    def g(x,t): return f(x[0],x[1])
    if limit is not None:
        ltime = (ltime or time)
        t = linspace(ltime[0],ltime[1],10000)
        x = odeint(g,array(limit),t)
        x = x[(x[:,0]>phase[0])*(x[:,1]>phase[1])*(x[:,0]<phase[2])*(x[:,1]<phase[3])]
        if lstart is None: lstart = lN//2
        plot(x[lstart:,0],x[lstart:,1],color='red')
    t = linspace(time[0],time[1],1000)
    x = odeint(g,array(start),t)
    subplot(122)
    if ma: mathaxes()
    plot(t,x[:,0]); plot(t,x[:,1])
```

```python
def f(x,y): return (-y,x)
odeanalyze(f,limit=(0.5,0.5))
```

```python
def f(x,y): return (x,0.5*y)
odeanalyze(f,limit=(0.5,0.5),lstart=0)
```

Stability criteria:

- look at the eigenvalues of $M$
- if they are both real, we have exponential growth
- if they are both imaginary, we have periodic solutions

# Real Pendulum

A pendulum also gives rise to harmonic oscillations. But the restoring force is proportional to the angle only for small $\theta$.

For large $\theta$, the restoring force is proportional to $\sin(x)$. This gives rise to a more interesting phase space portrait.



```python
def f(x,y): return (-y,sin(x))
odeanalyze(f,phase=(-8,-8,8,8),limit=(0.5,0.5))
```

# Lotka-Volterra Equation

Foxes vs hares:

$$ \frac{dH}{dt} = (a - bH - cF) H $$

$$ \frac{dF}{dt} = (-d + eH) F $$

Simple constants for Lotka-Volterra:

$$ \frac{dH}{dt} = (2 - H - F) H $$

$$ \frac{dF}{dt} = (-1 + H) F $$


```python
def f(H,F): return ((2-H-F)*H,(-1+H)*F)
odeanalyze(f,phase=(0,0,2,2),start=(0.1,0.1),limit=(0.5,0.5),lstart=0)
```

```python
def f(H,F): return (H*(1-F),-0.3*F*(1-H))
odeanalyze(f,phase=(0,0,2,2),start=(0.1,0.1),time=(0,70),limit=(0.8,0.8))
```

Note that close to equilibrium, we get harmonic oscillations.



```python
def f(H,F): return (H*(1-F),-F*(1-H))
odeanalyze(f,phase=(0.9,0.9,1.1,1.1),start=(1.05,1.05),time=(0,70),limit=(0.95,0.95))
```

# FitzHugh Nagumo Model

The FitzHugh-Nagumo modelis a simplified neuron model consisting of:

- a voltage-like variable $v$
- an inhibitory variable $w$
- a time-dependent current $I_e(t)$

$$ \dot{v} = v - v^3/3 - w + I_e(t) $$
$$ \tau \dot{w} = v - a - bw $$

```python
def fhm(v,w,a=0.0,b=0.0,tau=1.0,t=100,dt=1e-3,ie=lambda t:0.0):
    dv = (v-v**3/3.0-w+ie(t))*dt
    dw = ((v-a-b*w)/tau)*dt
    return dv,dw
odeanalyze(fhm,phase=(-4,-4,4,4),start=(0,-3),time=(0,30000),ma=1)
```

# van der Pol Oscillator and Limit Cycles

$$\dot x = y$$
$$\dot y = \mu(1-x^2) y-x.$$

```python
mu = 0.5
def vdp(x,y):
    dx = y
    dy = mu*(1-x**2)*y-x
    return dx,dy
odeanalyze(vdp,phase=(-6,-6,6,6),start=(1,1),time=(0,40),ma=1,limit=(4,4))
```

Limit cycle:

A limit cycle is an _isolated periodic orbit_.

# Chaos

So far, we have seen solutions to differential equations that...

- go off to infinity
- converge to a point
- converge to a limit cycle (periodic solution)

There is another class of solutions:

- aperiodic, bounded solutions

Such solutions may look approximately periodic, but they don't quite repeat exactly again. This is called a _chaotic solution_.

A simple form of chaos can be produced by adding time dependence to some non-linear oscillators.

Here is a _externally driven_ van der Pol oscillator. Note how the phase space trajectory is a limit cycle for the undriven case, but a dense ring for the driven case.



```python
mu = 0.5
def g(v,t): x,y = list(v); return y*(1+0.3*sin(1.73*t)**3),mu*(1-x**2)*y-x
t = linspace(0.0,1000.0,10000); x = odeint(g,(0.1,0.1),t); plot(x[5000:,0],x[5000:,1],color='blue')
def f(v,t): x,y = list(v); return y,mu*(1-x**2)*y-x
x = odeint(f,(0.1,0.1),t); plot(x[5000:,0],x[5000:,1],color='red')
```

We don't need external driving in order to obtain chaotic solutions (in fact, external driving is just another oscillator). We can produce the same result with a system of coupled non-linear differential equations. This system is called the _Lorenz system_ and occurs in climate modeling.



```python
def f(v,t,sigma=10,beta=8.0/3,rho=28.0): x,y,z = list(v); return sigma *( y - x ),x *( rho - z ) -y,x *y - beta * z
t = linspace(0.0,50.0,10000); x = odeint(f,(0.1,0.1,0.1),t); _=plot(t,x)
```

Slight differences in initial conditions lead to large deviations over time ("butterfly effect").



```python
t = linspace(0.0,50.0,10000); x1 = odeint(f,(0.1,0.1,0.1),t); x2 = odeint(f,(0.1001,0.1,0.1),t)
plot(t[1000:],x1[1000:,0]); plot(t[1000:],x2[1000:,0])
```

Chaotic solutions to different equations have _strange attractors_. These are neither limit cycles nor do they diverge, but they are _fractal sets_.



```python
from mpl_toolkits.mplot3d.axes3d import Axes3D
ax = gcf ().add_subplot(1,1,1,projection='3d')
ax.plot(x[:,0],x[:,1],x[:,2],color=(0 ,0 ,1 ,0.8) )
```

Here is the attractor projected down into 2D.



```python
plot(x[:,1],x[:,2])
```

chaotic solutions to differential equations:

- chaotic solutions to differential equations require external driving, $&ge;3$ dimensions and nonlinearities, or infinite dimensions
- chaotic solutions are aperiodic, bounded solutions
- chaotic solutions exist on strange attractors

# Chaos in Difference Equations

chaos in difference equations:

- the restrictions on chaotic systems don't exist for difference equations
- even simple difference equations can exhibit chaos, e.g. the logistic map

Logistic map:

$$x_{t+1} = r x (1-x) $$

```python
ms = []
x = 0.1
for i in range(100):
    x = 4*x*(1-x)
    ms.append(x)
plot(ms)
```

```python
def logistic_map(r,n=2048,iters=10000):
    result = zeros(n)
    x = 0.1
    for i in range(iters):
        result[int((len(result)-1)*x)] += 1
        x = r*x*(1-x)
        if x<0 or x>1: break
    return result/amax(result)
```

For 1D dynamical systems, we can show the phase space density while varying a single parameter.



```python
def expspace(lo,hi,n): return exp(linspace(log(lo),log(hi),n))
m = array([logistic_map(r,200) for r in linspace(0.0,4.0,200)])
imshow(m,cmap=cm.spectral)
```

The most interesting part of this map is near $r=4.0$, so let's focus on that.



```python
def expspace(lo,hi,n): return exp(linspace(log(lo),log(hi),n))
m = array([logistic_map(r) for r in (4.0-expspace(0.001,3.5,2048))])
imshow(m,cmap=cm.spectral,interpolation='nearest')
```

# Summary - Differential Equations

Classification of differential equations:

- ordinary vs partial
- linear vs non-linear

Phase space:

- plot flow field of derivatives
- look for nullclines
- look for asymptotic behavior
- look for limit cycles

Stability:

- in 1D, $f(x)=0$ and $f'(x)&lt;0$
- in 2D, $\dot x = 0$ and $\dot y = 0$ are stationary points
- stability given by the surrounding nullclines, eigenvalues of the Jacobian

Modeling:

- differential equations are usually continuous approximations to discrete systems
