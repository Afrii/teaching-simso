- body plans are generally created as a sequence of _segments_

General developmental sequence:

- start of with an undifferentiated tube
- divide the tube into segments
- differentiate each segment (attach legs, eyes, etc.)

Drosophila embryogenesis:

Mammals:

- cells divide, then migrate and differentiate
- each cell has a single nucleus
- communication / signaling between cells is complicated by cell membranes

Drosophila:

- the egg starts out with a single nucleus
- the nucleus divides 8 times, yielding 256 nuclei
- the 256 nuclei distribute themselves around the wall of the egg
- nuclei can send and receive signaling molecules that move via diffusion

```python
# Drosophila embryogenesis
fig("http://biology.kenyon.edu/courses/biol114/Chap13/droslifecycle.jpg")
```

Starting point:

- elongated container with 256 nuclei around the wall
- nuclei are _not_ differentiated
- we want to create a sequence of segments

How do we do that? What kind of signaling can create such patterns?

What happens next?

- naive approach: concentration sensing
- take the concentration of Caudal, or the ratio of Caudal to Hunchback
- divide the range into a bunch of disjoing regions
- "if ratio in $[1.7,2.4]$, then develop into segment 7"

That approach doesn't work; the signal is far too noisy.

How does it actually work?

- a series of _reaction-diffusion_ networks of morphogens
- morphogens enhance or repress both their own production and the production of other morphogens
- the whole network is similar to a complex network of predators, prey, and migration

Regulatory networks during morphogenesis:

Principles:

- interaction between morphogens establish disjoint regions (segments) within the organism
- each segment has an identifier based on the combination of morphogens expressed in them

High level view:

- analog signalling and non-linear interactions are used to establish a digital/binary code for morphogenesies
- regions are contiguous and have the correct topology (neighborhood relationships) with very high probability
